```python
# Remove unwanted whitespace from CLASS values to fix duplicate looking categories
df["CLASS"] = df["CLASS"].str.strip()
df["CLASS"].value_counts()
```

```python
# Remove ID column and duplicate patients (keep first occurrence)
df_clean = df.drop(columns=["ID"]).drop_duplicates(subset="No_Pation", keep="first")
df_clean.shape
```

```python
# Verify no duplicate patients remain
df_clean["No_Pation"].duplicated().sum()
```

```python
df_clean["CLASS"].value_counts()
```

```python
# Class distribution - percentages
df_clean["CLASS"].value_counts(normalize=True) * 100
```

```python
# Visualize class distribution - bar plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
df_clean["CLASS"].value_counts().plot(
    kind="bar", color=["#2ecc71", "#e74c3c", "#f39c12"]
)
plt.title("Class Distribution (N, Y, P)", fontsize=14, fontweight="bold")
plt.xlabel("Class", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Visualize class distribution - pie chart
plt.figure(figsize=(8, 6))
colors = ["#2ecc71", "#e74c3c", "#f39c12"]
df_clean["CLASS"].value_counts().plot(
    kind="pie", autopct="%1.1f%%", colors=colors, startangle=90
)
plt.title("Class Distribution Percentage", fontsize=14, fontweight="bold")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
# Check for class imbalance
class_counts = df_clean["CLASS"].value_counts()
imbalance_ratio = class_counts.max() / class_counts.min()
print(f"Imbalance ratio (max/min): {imbalance_ratio:.2f}")
```

```python
df_clean.describe()
```

```python
# Select only numerical features for analysis
numerical_cols = df_clean.select_dtypes(include=["number"]).columns.tolist()
numerical_cols.remove("No_Pation")  # Remove patient ID
print(f"Numerical features: {numerical_cols}")
```

```python
# Distribution plots for all numerical features
fig, axes = plt.subplots(4, 3, figsize=(15, 12))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols):
    df_clean[col].hist(bins=30, ax=axes[idx], edgecolor="black", alpha=0.7)
    axes[idx].set_title(f"{col} Distribution", fontweight="bold")
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# Box plots to detect outliers
fig, axes = plt.subplots(4, 3, figsize=(15, 12))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols):
    df_clean.boxplot(column=col, ax=axes[idx])
    axes[idx].set_title(f"{col} Boxplot", fontweight="bold")
    axes[idx].set_ylabel(col)

plt.tight_layout()
plt.show()
```

```python
# Count outliers using IQR method for each feature
for col in numerical_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_clean[(df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)]
    print(
        f"{col}: {len(outliers)} outliers ({len(outliers) / len(df_clean) * 100:.1f}%)"
    )
```

```python
# Gender distribution
df_clean["Gender"].value_counts()
```

```python
df_clean["Gender"] = df_clean["Gender"].str.upper()
df_clean["Gender"].value_counts()
```

```python
# Gender vs CLASS crosstab
pd.crosstab(df_clean["Gender"], df_clean["CLASS"], margins=True)
```

```python
# Correlation matrix
import seaborn as sns
correlation_matrix = df_clean[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    square=True,
    linewidths=1,
    cbar_kws={"shrink": 0.8},
)
plt.title("Correlation Matrix of Numerical Features", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
# Simple violin plots for key features vs CLASS
key_features = ["HbA1c", "BMI", "AGE", "Chol"]

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()

for idx, col in enumerate(key_features):
    sns.violinplot(
        data=df_clean,
        x="CLASS",
        y=col,
        hue="CLASS",
        legend=False,
        ax=axes[idx]
    )
    axes[idx].set_title(f"{col} by CLASS")

plt.tight_layout()
plt.show()
```

```python
# Check which features might have unusual distributions
for col in numerical_cols:
    unique_count = df_clean[col].nunique()
    print(
        f"{col}: {unique_count} unique values, min={df_clean[col].min()}, max={df_clean[col].max()}"
    )
```

```python
# Investigate extreme Cr values
extreme_cr = df_clean[df_clean["Cr"] > 100]
print(f"Patients with Cr > 100: {len(extreme_cr)}")
extreme_cr[["AGE", "Cr", "Urea", "HbA1c", "CLASS"]].sort_values(
    "Cr", ascending=False
).head(10)
```

```python
# Investigate very low cholesterol values
low_chol = df_clean[df_clean["Chol"] < 2]
print(f"Patients with Chol < 2: {len(low_chol)}")
low_chol[["AGE", "Chol", "HDL", "LDL", "VLDL", "CLASS"]].head(10)
```
